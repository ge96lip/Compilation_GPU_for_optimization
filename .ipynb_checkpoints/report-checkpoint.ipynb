# Exercise 1 - Gauss-Seidel for Poisson Solver


### Task 1.1: 
Develop the Gauss-Seidel solver with Python List, array, or NumPy. Plot the performance varying the grid size. We noticed that as the grid size increased, so did the execution time. 

#### code

```python
def gauss_seidel(f, num_iterations=1000):
    """
    Perform Gauss-Seidel iterations to solve the 2D Poisson equation.

    Parameters:
    - f: 2D NumPy array representing the grid.
    - num_iterations: Number of iterations to perform.

    Returns:
    - Updated 2D array after Gauss-Seidel iterations.
    """
    newf = f.copy()
    n, m = newf.shape

    for _ in range(num_iterations):
        for i in range(1, n-1):
            for j in range(1, m-1):
                newf[i, j] = 0.25 * (newf[i, j+1] + newf[i, j-1] +
                                     newf[i+1, j] + newf[i-1, j])
    return newf

def run_performance_test(grid_sizes, num_iterations=1000):
    """
    Run the Gauss-Seidel solver for different grid sizes and measure execution time.

    Parameters:
    - grid_sizes: List of grid sizes to test.
    - num_iterations: Number of iterations for Gauss-Seidel method.

    Returns:
    - Dictionary mapping grid size to execution time.
    """
    performance = {}

    for N in grid_sizes:
        print(f"Running Gauss-Seidel for grid size {N}x{N}...")
        f = np.random.rand(N, N)  # Initialize grid with random numbers
        f[0, :], f[-1, :], f[:, 0], f[:, -1] = 0, 0, 0, 0  # Set boundary conditions

        start_time = time.time()
        gauss_seidel(f, num_iterations)
        elapsed_time = time.time() - start_time

        performance[N] = elapsed_time
        print(f"Grid size {N}x{N} took {elapsed_time:.4f} seconds")

    return performance

def run_profiler():

    # Define grid sizes
    grid_sizes = [10, 20, 50, 100] # could include 200

    # Run performance test
    performance_data = run_performance_test(grid_sizes)

    # Plot performance results
    plt.figure(figsize=(8, 6))
    plt.plot(performance_data.keys(), performance_data.values(), marker='o', linestyle='-')
    plt.xlabel("Grid Size (N x N)")
    plt.ylabel("Execution Time (seconds)")
    plt.title("Gauss-Seidel Solver Performance")
    plt.grid()
    # plt.show()
```

#### results

```python
run_profiler()
```

# Task 1.8 
Save the newgrid matrix as an hdf5 file using h5py

### Code

```python
import h5py

def gauss_seidel_hdf5(f, num_iterations=1):
    hf = h5py.File("newgrid.hdf5", 'a')
    dset = hf.create_dataset("n_100",(100,100))
    newf = f.copy()
    n, m = newf.shape

    for _ in range(num_iterations):
        for i in range(1, n-1):
            for j in range(1, m-1):
                val = 0.25 * (newf[i, j+1] + newf[i, j-1] +
                                     newf[i+1, j] + newf[i-1, j])
                newf[i, j] = val
                cell = f"cell_{i}_{j}"
                # dset.create("n_1000")
                hf["n_100"].attrs[cell] = val

    print("File datasets: ", list(hf.keys()))
    print("Dataset shape:", dset.shape)
    print("Most recent attribute: ", cell)
    print("Most recently added cell value: ", dset.attrs[cell])

    del dset
    hf.close() 

    return newf
```

### Results

```python
N = 100
f = np.random.rand(N, N)  # Initialize grid with random numbers
f[0, :], f[-1, :], f[:, 0], f[:, -1] = 0, 0, 0, 0  # Set boundary conditions
newf_for_hdf5 = gauss_seidel_hdf5(f)
```

# Bonus Exercise - Fast Fractal Fun with Cython and GPUs

## Mandelbrot code

```python
import numpy as np
import matplotlib.pyplot as plt

def mandelbrot(c, max_iter=1000):
    """Computes the number of iterations before divergence."""
    z = 0
    for n in range(max_iter):
        if abs(z) > 2:
            return n
        #print(z, ", ", type(z))
        #print(c, ", ", type(c))
        z = z*z + c
    return max_iter

def mandelbrot_set(width, height, x_min, x_max, y_min, y_max, max_iter=100):
    """Generates the Mandelbrot set image."""
    x_vals = np.linspace(x_min, x_max, width)
    y_vals = np.linspace(y_min, y_max, height)
    image = np.zeros((height, width))

    for i in range(height):
        for j in range(width):
            c = complex(x_vals[j], y_vals[i])
            image[i, j] = mandelbrot(c, max_iter)

    return image

# Parameters
width, height = 1000, 800
x_min, x_max, y_min, y_max = -2, 1, -1, 1

# Generate fractal
start = time.time()
image = mandelbrot_set(width, height, x_min, x_max, y_min, y_max)
elapsed = time.time() - start
print("Elapsed time (original): ", elapsed)
# Display
plt.imshow(image, cmap='inferno', extent=[x_min, x_max, y_min, y_max])
plt.colorbar()
plt.title("Mandelbrot Set")
plt.show()
```

## B.2 - CuPy

### CP Code Version

```python
import cupy as cp
def mandelbrot_set_cp(width, height, x_min, x_max, y_min, y_max, max_iter=100):
    """Generates the Mandelbrot set image."""
    x_vals = cp.linspace(x_min, x_max, width)
    y_vals = cp.linspace(y_min, y_max, height)
    x_grid,y_grid = cp.meshgrid(x_vals, y_vals)
    complex_grid = x_grid + y_grid * 1j
    mandelbrot(complex_grid)

    image = cp.zeros((height, width))
    image = mandelbrot(complex_grid, max_iter)

    return image

def mandelbrot(complex_grid, max_iter=100):
    """Computes the number of iterations before divergence."""
    ret_grid = cp.zeros((height, width))
    z_grid = cp.zeros((height, width), dtype=cp.complex128)
    two_grid = cp.full((height, width), 2)
    for n in range(max_iter):
        # boolean masking
        abs_grid = cp.absolute(z_grid)
        mask_grid = cp.less_equal(abs_grid, two_grid)
        # return n
        cp.putmask(ret_grid, mask_grid, n)
        # updating z
        cp.putmask(z_grid, mask_grid, z_grid * z_grid + complex_grid)
    return ret_grid




```

### Testing

```python
import time
width, height = 1000, 800
x_min, x_max, y_min, y_max = -2, 1, -1, 1

# Generate fractal
start = time.time()
image = mandelbrot_set_cp(width, height, x_min, x_max, y_min, y_max)
elapsed = time.time() - start
print("Elapsed time (original): ", elapsed)
# Display
plt.imshow(image.get(), cmap='inferno', extent=[x_min, x_max, y_min, y_max])
plt.colorbar()
plt.title("Mandelbrot Set")
plt.show()
```

The original Mandelbrot code took: 3.5875 seconds. After our optimizations with CuPy, we ran the code in: 0.8844 seconds. The CuPy optimizations with meshgrid and boolean masking gave us an increase of 2.77031 seconds!
